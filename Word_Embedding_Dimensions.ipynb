# **GENSIM**

```python
new_text=df.review.apply(gensim.utils.simple_preprocess)
new_text
```

# **Embedding_Dim** =50

```python
Embedding_Dim1 =50
model1 = gensim.models.Word2Vec(sentences = new_text, size = Embedding_Dim1, workers = 4)
#if sg=1 skipgram 
v = list(model1.wv.vocab)
print(' Vocabulary Size.. %d' % len(v))
```

```python
model1.wv.most_similar('good')
```

```python
model1.wv.most_similar_cosmul(positive=['actor','boy'], negative=['girl'])
```

```python
model1.wv.most_similar_cosmul(positive=['india','delhi'], negative=['england'])
```

```python
model1.wv.most_similar_cosmul(positive=['king','man'], negative=['women'])
```

```python
model1.wv.most_similar('actor')
```

```python
model1.wv.most_similar('king')
```

```python
model1.wv.doesnt_match('king boy water'.split())
```

```python
model1.wv.doesnt_match('king queen water'.split())
```
